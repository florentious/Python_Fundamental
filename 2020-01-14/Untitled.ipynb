```python
def minimumBribes(q) :
    # 역으로 저걸 sort해가는 최소값인가
    count = 0
    sort = []
    for i in range(1,len(q)+1) :
        sort.append(i)
    
    # check 'Too chaotic'
    for i,v in enumerate(q) :
        if (v-1) - i > 2 :
            print("Too chaotic")
            return
    # check don't chaotic : bubble sort
    for i in range(len(q)-1) :
        for j in range(len(q)-1) :
            if q[j] > q[j+1] :
                tmp = q[j]
                q[j] = q[j+1]
                q[j+1] = tmp
                count+=1
                
        if sort == q :
            break
    print(count)
    return
            
    
```

```python
def getMid(arr) :
    arr.sort()
    if len(arr) % 2 == 1 :
        return(arr[len(arr)//2])
    else :
        return((arr[len(arr)//2] + arr[(len(arr)//2)-1])/2)
```

```python
def activityNotifications(expenditure, d):
    count = 0

    for i in range(d, len(expenditure)) :
        mid = getMid(expenditure[i-d:i])
        if mid*2 <= expenditure[i] :
            count+=1

    return(count)
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://www.rottentomatoes.com/"
html = urlopen(url)
source = html.read() # 바이트코드 type으로 소스를 읽는다.
html.close() # urlopen을 진행한 후에는 close를 한다.

soup = BeautifulSoup(source, "html5lib") # 파싱할 문서를 BeautifulSoup 클래스의 생성자에 넘겨주어 문서 개체를 생성, 관습적으로 soup 이라 부름
table = soup.find(id="Top-Box-Office")
movies = table.find_all(class_="middle_col")

for movie in movies:
    title = movie.get_text()
    print(title, end=' ')
    link = movie.a.get('href')
    url = 'https://www.rottentomatoes.com' + link
    print(url)
```
