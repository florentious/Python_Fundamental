```python
pd.pivot_table(data,index=['Manager','Rep'],values=['Price'],aggfunc=[np.mean,np.sum,len])
```

```python
pd.pivot_table(data,index=['Manager','Rep'],values=['Price'],columns=['Product'],aggfunc=[np.sum],fill_value=0)


# 내부에 있던 자료가 들어갈 자리가 없는 경우에는 NaN값이 들어갈 수 있다.
# fill_value = 0 옵션으로 None 값에 대해서 0으로 초기화해서 처리함
```

```python
pd.pivot_table(data,index=['Manager','Rep','Product'],values=['Price','Quantity'],aggfunc=[np.sum],fill_value=0)

# 상황에따라서 좌랑 우측의 밸류로 초기화해서 표를 볼 수 있다
```

```python
pd.pivot_table(data,index=['Manager','Rep','Product'],
               values=['Price','Quantity'],aggfunc=[np.sum,np.mean],fill_value=0, margins=True)

# margins = True : 총계를 찍어주는 것
# aggfunc에 numpy 함수 외에는 잘 안먹는것을 알 수 있다.
```

```python
pd.pivot_table(data,index=['Manager','Rep','Product'],
               values=['Price','Quantity'],aggfunc={'Price':np.sum , 'Quantity':len},fill_value=0, margins=True)

# 해당 컬럼에 대해서 각각 함수를 지정해서 쓸수 있는 것
```

```python
pd.pivot_table(data,index=['Manager','Rep'],columns=['Product'],
               values=['Price','Quantity'],aggfunc={'Quantity':len, 'Price':[np.sum, np.mean]},fill_value=0)

# margins 들어가면 에러남 왜그럼?
```

```python
table = pd.pivot_table(data,index=['Manager','Rep'],columns=['Product'], values=['Price','Quantity'],aggfunc={'Quantity':len, 'Price':[np.sum, np.mean]},fill_value=0)

table.query('Manager == ["Debra Henley"]')
```
